```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
city_day_data = pd.read_csv('C:/Air-Pollution-Prediction-System/data/city_day.csv')

# Convert 'Date' to datetime object
city_day_data['Date'] = pd.to_datetime(city_day_data['Date'])

# Extract month and year from 'Date' for analysis
city_day_data['Month'] = city_day_data['Date'].dt.month
city_day_data['Year'] = city_day_data['Date'].dt.year

# Group by city and month, calculate average AQI
avg_aqi_by_city_month = city_day_data.groupby(['City', 'Month'])['AQI'].mean().reset_index()

# Plotting the average AQI by city for each month
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_aqi_by_city_month, x='Month', y='AQI', hue='City', marker='o')
plt.title('Average AQI by City and Month (Daily Data)')
plt.xlabel('Month')
plt.ylabel('Average AQI')
plt.xticks(range(1, 13))
plt.grid(True)
plt.show()
```

```python
# Load dataset
city_hour_data = pd.read_csv('C:/Air-Pollution-Prediction-System/data/city_hour.csv')

# Convert 'Datetime' to datetime object
city_hour_data['Datetime'] = pd.to_datetime(city_hour_data['Datetime'])

# Extract hour from 'Datetime' for analysis
city_hour_data['Hour'] = city_hour_data['Datetime'].dt.hour

# Group by city and hour, calculate average AQI
avg_aqi_by_city_hour = city_hour_data.groupby(['City', 'Hour'])['AQI'].mean().reset_index()

# Plotting the average AQI by city and hour
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_aqi_by_city_hour, x='Hour', y='AQI', hue='City', marker='o')
plt.title('Average AQI by City and Hour (Hourly Data)')
plt.xlabel('Hour of the Day')
plt.ylabel('Average AQI')
plt.xticks(range(0, 24))
plt.grid(True)
plt.show()
```

```python
# Load dataset
station_day_data = pd.read_csv('C:/Air-Pollution-Prediction-System/data/station_day.csv')

# Convert 'Date' to datetime object
station_day_data['Date'] = pd.to_datetime(station_day_data['Date'])

# Extract month and year for analysis
station_day_data['Month'] = station_day_data['Date'].dt.month
station_day_data['Year'] = station_day_data['Date'].dt.year

# Group by station and month, calculate average AQI
avg_aqi_by_station_month = station_day_data.groupby(['StationId', 'Month'])['AQI'].mean().reset_index()

# Plotting the average AQI by station for each month
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_aqi_by_station_month, x='Month', y='AQI', hue='StationId', marker='o')
plt.title('Average AQI by Station and Month (Daily Data)')
plt.xlabel('Month')
plt.ylabel('Average AQI')
plt.xticks(range(1, 13))
plt.grid(True)
plt.show()
```

```python
# Load dataset
station_hour_data = pd.read_csv('C:/Air-Pollution-Prediction-System/data/station_hour.csv')

# Convert 'Datetime' to datetime object
station_hour_data['Datetime'] = pd.to_datetime(station_hour_data['Datetime'])

# Extract hour from 'Datetime' for analysis
station_hour_data['Hour'] = station_hour_data['Datetime'].dt.hour

# Group by station and hour, calculate average AQI
avg_aqi_by_station_hour = station_hour_data.groupby(['StationId', 'Hour'])['AQI'].mean().reset_index()

# Plotting the average AQI by station and hour
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_aqi_by_station_hour, x='Hour', y='AQI', hue='StationId', marker='o')
plt.title('Average AQI by Station and Hour (Hourly Data)')
plt.xlabel('Hour of the Day')
plt.ylabel('Average AQI')
plt.xticks(range(0, 24))
plt.grid(True)
plt.show()
```

```python
# Load dataset
stations_data = pd.read_csv('C:/Air-Pollution-Prediction-System/data/stations.csv')

# Displaying station information
print(stations_data.head())

# Plotting the number of stations by city
stations_per_city = stations_data['City'].value_counts().reset_index()
stations_per_city.columns = ['City', 'Number of Stations']

plt.figure(figsize=(12, 6))
sns.barplot(data=stations_per_city, x='City', y='Number of Stations')
plt.title('Number of Stations by City')
plt.xlabel('City')
plt.ylabel('Number of Stations')
plt.xticks(rotation=90)
plt.show()

# Plotting the number of stations by state
stations_per_state = stations_data['State'].value_counts().reset_index()
stations_per_state.columns = ['State', 'Number of Stations']

plt.figure(figsize=(12, 6))
sns.barplot(data=stations_per_state, x='State', y='Number of Stations')
plt.title('Number of Stations by State')
plt.xlabel('State')
plt.ylabel('Number of Stations')
plt.xticks(rotation=90)
plt.show()
```
